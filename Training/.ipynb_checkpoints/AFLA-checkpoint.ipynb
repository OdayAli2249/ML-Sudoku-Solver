```python
import os
import requests
import io #codecs


# Text version of https://kilgarriff.co.uk/Publications/2005-K-lineer.pdf
if os.path.isfile('plato1.txt'):
    with io.open('plato1.txt', encoding='utf8') as fin:
        text = fin.read()
        
###################

# tokenizer = RegexpTokenizer(r'\"(\w+|\$[\d\.]+|\S+|\s+)+\"')
# tokenizer = RegexpTokenizer(r'^\“(\s*\S+\s+)+\”$')
quotes = re.findall(r'“.*”',text)
#regtext = tokenizer.tokenize(text)
#print(tokenizer.tokenize(text))

###################

stop_words = set(stopwords.words('english'))
Punction_set = [',','.','"',':','?','','“','”']
filtered_Quotes = []

# Tokenize the text.
tokenizedQuotes = []
for quote in quotes:
    after = [list(map(str.lower, word_tokenize(sent))) 
                  for sent in sent_tokenize(quote)]
    for temp in after:
        tokenizedQuotes.append(temp)
        
for temp in tokenizedQuotes:
    #temp1 = [w for w in temp if (not w in stop_words and not w in Punction_set)]
    temp1 = [w for w in temp if not w in Punction_set]
    filtered_Quotes.append(temp1)

allwords = []
for temp in filtered_Quotes:
    for w in temp:
        allwords.append(w)
        
        
allwords1 = []
for temp in tokenizedQuotes:
    for w in temp:
        allwords1.append(w)
```

```python
from nltk.tokenize.treebank import TreebankWordDetokenizer

detokenize = TreebankWordDetokenizer().detokenize

def generate_sent(model, num_words, random_seed=42):
    """
    :param model: An ngram language model from `nltk.lm.model`.
    :param num_words: Max no. of words to generate.
    :param random_seed: Seed value for random.
    """
    content = []
    for token in model.generate(num_words, random_seed=random_seed):
        if token == '<s>':
            continue
        if token == '</s>':
            break
        content.append(token)
    return detokenize(content)
```
